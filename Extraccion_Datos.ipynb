```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('Brasil Rio de Janeiro.csv' , encoding='latin-1')
df
```

FITRO OR

```python
#Los registros de 5 host name diferentes
#FILTRO OR
filtro1 = df[(df["host_name"] == 'Lucas') | (df["host_name"] == 'Robert') | (df["host_name"] == 'Antonia Eridete') |  (df["host_name"] == 'Paulo') | (df["host_name"] == 'Celina')]
filtro1
```

FILTRO COMPARACION OPERADOR LOGICO

```python
#los registros de los host que se hayan unido a airbnb antes del año 2019
#FILTRO COMPARACION 
#compara que los anfitriones que se unieron a Airbnb antes del 1 de enero de 2019
#verifica si la fecha de unión del anfitrión es anterior a esa fecha
filtro2 = df[df["host_since"] < '2019-01-01']
filtro2
```

FILTRO OR Y COMPARACION 

```python
#los registros de los host que resondan a mas tardar en un dia
#filtro por comparacion y or
#responden en 1 día o menos verifica si el tiempo de respuesta del anfitrión es igual o inferior a un dia solo los que cumplan la condicion
filtro3 = df[(df["host_response_time"] == 'within a day') | (df["host_response_time"] == 'within a few hours') | (df["host_response_time"] == 'within an hour') ]
filtro3
```

FILTRO COMPARACION OPERADOR LOGICO

```python
#los registros que tengan al menos 5 accomodates
#filtro por comparacion
#la cantidad de personas que un alojamiento puede acommodates es de al menos 5
#solo aquellos que cumplen 
filtro4=df[df["accommodates"] >= 5]
filtro4
```

FILTRO OR

```python
#registros de los tipos de cuarto "Entire home" y "Private room"
#filtro OR 
#filas donde el tipo de cuarto es "Entire home" o "Private room" cualquiera de 
#las dos que cumpla lo selecciona 
filtro5 = df[(df["room_type"] == "Entire home/apt") | (df["room_type"] == "Private room")]
filtro5
```

FILTRO COMPARACION OPERADOR LOGICO

```python
#filtro que "a lo mucho cuenten con 4 camas"
#filtro por comparacion
#para que seleccione endonde el numero de camas es menor o igual a 4 
filtro6 = df[df["beds"] <= 4]
filtro6
```

FILTRO AND & Y OR 

```python
#los registros que tengan un baño o que sean super host pero la condicion es que todos sean instant_bookable
#filtro 7
#filtro and & y Or | seleccionen los registros donde la propiedad sea instant_bookable para seleccionar los registros que tengan más de 0 baños 
#o aquellos donde el anfitrión sea un superhost 
filtro7 = df[(df["instant_bookable"] == 't') & ((df["bathrooms"] == 1) | (df["host_is_superhost"] == 't'))]
filtro7
```

FILTRO COMPARACION Y AND

```python
#los registros cuyo "review_scores_cleanliness" sea menor a 3 y "review_scores_raiting" sea mayor de 3
#filtro de comparación
#sobre ambas columnas usando < y & se deben cumplir ambas condiciones en ambas filas
#para que aparezcan
filtro8 = df[(df["review_scores_cleanliness"] < 3) & (df["review_scores_rating"] > 3)]
filtro8
```

```python
filtro1.to_csv('filtro1_OR.csv')
filtro2.to_csv('filtro2_comparacion.csv')
filtro4.to_csv('filtro4_comparacion.csv')
filtro5.to_csv('filtro5_OR.csv')
filtro6.to_csv('filtro6_comparacion.csv')
filtro7.to_csv('filtro7_and_Or.csv')
filtro8.to_csv('filtro8_comparacion.csv')
```
