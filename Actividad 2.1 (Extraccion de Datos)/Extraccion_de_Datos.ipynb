```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
df
```

```python
df.info()
```

```python
#filtro por objeto
filtro1=df[(df["NOMBRE_VENDEDOR"]== "DIEGO ARMANDO LEON DURAN")]
filtro1
#extrae y muestra las filas donde el vendedor es DIEGO 
```

```python
filtro1.to_csv('filtro1_objeto.csv')
```

```python
#filtro por filas
filtro2=df.iloc[76:82,:] #filas consecutivas
#filtro2= df.iloc[[2, 5] , :] #filas no consecutivas
filtro2
# : significa que imprimira todas las columnas 
#no consecutivas: [[2, 5], :]
#consecutivas: [2:6,:]
```

```python
filtro2.to_csv('filtro2_filas.csv')
```

```python
#filtro por columnas
filtro3=df.iloc[ : , [3,8,13]]
filtro3
# : todas las filas , [3,8,13] que columnas contando 0
```

```python
filtro3.to_csv('filtro3_columnas.csv')
```

```python
#filtro por comparacion
filtro7=df[df["MARGEN_TOTAL_CALCULADO"]>8000]
filtro7
#obtener solo las filas la columna "MARGEN_TOTAL_CALCULADO" tiene un valor mayor a 8000
```

```python
filtro7.to_csv('filtro7_comparacion.csv')
```

```python
#filtro y conectores logicos
filtro8=df[(df["COSTO_UNITARIO"]<3000) & (df["NOMBRE_VENDEDOR"] == "AARON NUÑO CORDOVA")]
filtro8
#amabas condiciones deben cumplirse para mostrarse "COSTO_UNITARIO" es menor a 3000
#"NOMBRE_VENDEDOR" es igual a "AARON NUÑO CORDOVA"
```

```python
filtro8.to_csv('filtro8_conectores.csv')
```

```python
#filtro o
filtro9=df[(df["CANT"] > 10) | (df["FECHA_DOC"] == "2022-01-05")]
filtro9
#La columna "CANT" tiene un valor mayor a 10
#La columna "FECHA_DOC" es igual a "2022-01-05"
```

```python
filtro9.to_csv('filtro9_o.csv')
```

```python
#filtro not
filtro10=df[~(df["CVE_ART"]=="RP00230")]
filtro10
#elimina las filas que tienen el valor especifico selecionado de la variable 
#CVE_ART 
```
