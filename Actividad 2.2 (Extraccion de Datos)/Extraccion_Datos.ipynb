```python
#librerias necesarias 
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#cargar el excel 
df=pd.read_excel('Datos de Facturación.xlsx')
df
```

```python
#datos que tiene el DataFrame
df.info()
```

```python
#CVE_CLPV      (valores: 1000 a 2000)
#filtro y de la variabale CVE_CLPV convertimos con pd.to_numeric a tipo numerico
#errors="coerce" por si no hay numeros poner NA y filtra solo las filas qeu esten entre
#1000 y 2000 ya que habia problemas porque el valor original es de tipo object
df["CVE_CLPV"]=pd.to_numeric(df["CVE_CLPV"], errors="coerce")
filtro1 = df[(df["CVE_CLPV"] >= 1000) & (df["CVE_CLPV"] <= 2000)]
filtro1
```

```python
#CVE_VEND     (Todas las claves excepto “5” y “4”)
#filtro not y filtro or para solo extraer los necesario y no 5 y 4 
filtro2=df[~(df["CVE_VEND"] == 5.00 ) | (df["CVE_VEND"] != 4.00)]
filtro2
```

```python
#FECHA_ENT   (Fechas de “28/02/2022”)
#filtro por objeto, convertimos FECHA_ENT en formato de fecha y hora datatime para que
#dayfirst=True asegura que el día se interprete primero para filtrar solo filas 8/02/2022 (sin considerar la hora)
df["FECHA_ENT"] = pd.to_datetime(df["FECHA_ENT"], format="%d/%m/%Y/%H:%M", errors="coerce", dayfirst=True)
filtro3 = df[df["FECHA_ENT"] == pd.to_datetime("28/02/2022")]
filtro3
```

```python
#CAN_TOT       (Cantidades menores a 5951.7)   o  STATUS (“E”)
#filtro o ya que necesitamos extraer datos cantidades menores a 5951.7 o con
#status E algunos de esos dos 
filtro4=df[(df["CAN_TOT"] < 5951.7) | (df["STATUS"] == "E")]
filtro4
```

```python
#Solo las columnas: CVE_DOC, FECHA_ENT, FECHA_VEN y CAN_TOT
#filtro por columnas con indices tomando en cuenta el 0 primero
filtro5=df.iloc[ : , [0,6,7,8]] #Columnas no consecutivas
filtro5
```

```python
#Solo las filas de 7001-7099
#filtro por filas para extraer desde 7001 hasta 7099 
filtro6=df.iloc[7001:7099,:]
filtro6
```

```python
#Index= CVE_VEND (valores: 1, 2)  (columna: FECHAELAB)
#cargamos el archivo y especificamos que la columna número 3 (la cuarta columna, ya que la numeración 
#empieza desde 0) debe usarse como índice del DataFrame
df1=pd.read_excel('Datos de Facturación.xlsx', index_col=3)
df1
```

```python
#Index= CVE_VEND (valores: 1, 2)  (columna: FECHAELAB)
#filtro filas vs columnas en donde los valores de CVE_VEND deben ser 1 y 2
#y la columna de FECHAELAB
filtro7=df1.loc[[1.0, 2.0], ["FECHAELAB"]]
filtro7
```
